```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
%matplotlib inline
```

```python
#importing training dataset and displaying
train_data=pd.read_csv('train.csv')
train_data.head()
```

```python
#importing test dataset and displaying
test_data=pd.read_csv('test.csv')
test_data.head()
```

```python
# dealing with null values in training and test dataset for continious features 
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean') #Replacing all missing values with the mean of particular column
imp_mean.fit(train_data[['A2']])
train_data['A2']= imp_mean.transform(train_data[['A2']]).ravel()
imp_mean.fit(train_data[['A14']])
train_data['A14']= imp_mean.transform(train_data[['A14']]).ravel()
imp_mean.fit(test_data[['A2']])
test_data['A2']= imp_mean.transform(test_data[['A2']]).ravel()
imp_mean.fit(test_data[['A14']])
test_data['A14']= imp_mean.transform(test_data[['A14']]).ravel()
```

```python
# dealing with null values for categorical features
imp_mean = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp_mean.fit(train_data[['A1']])
train_data['A1']= imp_mean.transform(train_data[['A1']]).ravel()
imp_mean.fit(train_data[['A4']])
train_data['A4']= imp_mean.transform(train_data[['A4']]).ravel()
imp_mean.fit(train_data[['A5']])
train_data['A5']= imp_mean.transform(train_data[['A5']]).ravel()
imp_mean.fit(train_data[['A6']])
train_data['A6']= imp_mean.transform(train_data[['A6']]).ravel()
imp_mean.fit(train_data[['A7']])
train_data['A7']= imp_mean.transform(train_data[['A7']]).ravel()
imp_mean.fit(test_data[['A1']])
test_data['A1']= imp_mean.transform(test_data[['A1']]).ravel()
imp_mean.fit(test_data[['A4']])
test_data['A4']= imp_mean.transform(test_data[['A4']]).ravel()
imp_mean.fit(test_data[['A5']])
test_data['A5']= imp_mean.transform(test_data[['A5']]).ravel()
imp_mean.fit(test_data[['A6']])
test_data['A6']= imp_mean.transform(test_data[['A6']]).ravel()
imp_mean.fit(test_data[['A7']])
test_data['A7']= imp_mean.transform(test_data[['A7']]).ravel()
```

```python
# Set Train data and Test data
X_train=train_data.drop('A16',axis=1) #AS 'A16' is the target column so we remove from X_train
Y_train=train_data['A16']         #Set 'A16' as the dependent variable Y_train
X_test=test_data.drop('A16',axis=1)
Y_test=test_data['A16']
```

```python
# Check first 10 rows of X_train dataset
X_train.head(10)
```

```python
#Check first 10 rows of Y_Train
Y_train.head(10)
```

```python
# scaling the dataset
from sklearn.preprocessing import StandardScaler
X_train=StandardScaler().fit_transform(X_train)  
X_test=StandardScaler().fit_transform(X_test)
```

```python
# building models

def models(X_train,Y_train):
        #logistic regression
        from sklearn.linear_model import LogisticRegression
        log_model=LogisticRegression(random_state=0)
        log_model.fit(X_train,Y_train)
        
        #K-Nearest Neighbor
        from sklearn.neighbors import KNeighborsClassifier
        knn=KNeighborsClassifier(n_neighbors=5)
        knn.fit(X_train,Y_train)
        
        #Decision Tree
        from sklearn.tree import DecisionTreeClassifier
        dtree=DecisionTreeClassifier(random_state=0,criterion="entropy")
        dtree.fit(X_train,Y_train)
        
        #Random Forest
        from sklearn.ensemble import RandomForestClassifier
        rfc=RandomForestClassifier(random_state=0,criterion="entropy",n_estimators=15)
        rfc.fit(X_train,Y_train)
        
        print('[0]logistic regression accuracy:',log_model.score(X_train,Y_train))
        print('[1]K-Nearest Neighbor accuracy:',knn.score(X_train,Y_train))
        print('[2]Decision tree accuracy:',dtree.score(X_train,Y_train))
        print('[3]Random forest accuracy:',rfc.score(X_train,Y_train))
        
        return log_model,knn,dtree,rfc
```

```python
# compiling model
model=models(X_train,Y_train.values.ravel())
```

```python
# Evaluating Performance of the model
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

for i in range(len(model)):
    predictions=model[i].predict(X_test)
    print("Model",i)
    print(confusion_matrix(Y_test,predictions))
    print(classification_report(Y_test.values.ravel(),predictions))
    print('Accuracy : ',accuracy_score(Y_test.values.ravel(),predictions))
```

```python
# prediction of random-forest since from the evaluation of perfromance for each model above we get most accurate result in Random Forest
predictions=model[3].predict(X_test)
print('Predicted values:')
print(predictions)
print('Actual values:')
print(Y_test.values.ravel())
plt.scatter(Y_test.values.ravel(),predictions)
```
