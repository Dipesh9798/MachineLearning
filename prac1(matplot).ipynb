```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.show()
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
plt.plot(x,y,'r-')
#plt.show()
plt.subplot(1,2,1)
plt.plot(x,y,'b-')
plt.subplot(1,2,2)
plt.plot(y,x,'r-')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title')
axes.plot(x,y)
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
```
